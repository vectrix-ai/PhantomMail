## Create shipments

```python
import random

customer_ids = [
    "VANDERMEER-001",
    "BRADWICK-001",
    "NORDVIK-001",
    "VERMEULEN-001",
    "CASTELLANO-001",
    "RICHTER-001",
    "KOWALSKI-001",
    "MORETTI-001",
    "DEBRUIJN-001",
    "STEINBACH-001",
    "PEMBERTON-001",
    "DUBOIS-001",
    "BALTIKA-001",
    "KELLER-001",
    "COPPENS-001",
    "DELACROIX-001",
    "TERMINALNOORD-001",
    "VANHOUTEN-001",
    "PSZ-001",
    "APM-001"
]

url = 'https://mpl-backend-dev.vectrix.app/api/v1/shipments/'
shipment_responses = []

for response in responses:
    # Select a random customer ID between 1 and 19
    random_customer_id = random.choice(customer_ids)
    
    payload = {
        "message_id": response.json()['message_id'],
        "customer_id": random_customer_id,
        "customer_reference": "PO-2024-001",
        "requested_price": random.randint(800, 1500),
        "currency": "EUR",
        "weight": random.randint(500, 1350),
        "weight_unit": "kg",
        "volume": 2.5,
        "volume_unit": "m3",
        "pallet_count": random.randint(1, 5),
        "length": random.randint(2, 12),
        "length_unit": "m",
        "status": "waiting",
        "external_ids": {
            "customer_order_id": "ORD-12345",
            "tracking_number": "TRK-98765"
        },
        "template_document": "standard_transport",
        "external_metadata": {
            "special_instructions": "Handle with care",
            "goods_description": "Electronic components"
        },
        "parameters": {
            "insurance_required": True,
            "temperature_controlled": False
        }
    }

    response = requests.post(url, json=payload)
    print(f"Shipment response for message {response.json()['message_id']} with customer {random_customer_id}: {response.status_code} - {response.text}")
    shipment_responses.append(response)
```

## Create Stops

```python
import pandas as pd

# Load the real EU addresses from CSV
eu_addresses_df = pd.read_csv('eu_addresses.csv')
print(f"Loaded {len(eu_addresses_df)} real EU addresses")
eu_addresses_df.head()
```

```python
url = 'https://mpl-backend-dev.vectrix.app/api/v1/stops/'

eu_locales = ['en_GB', 'de_DE', 'fr_FR', 'it_IT', 'es_ES', 'nl_NL', 'pl_PL', 'pt_PT', 'sv_SE', 'da_DK']
fake = Faker(eu_locales)

for shipment_response in shipment_responses:
    # Select random pickup address from real EU addresses
    pickup_address = eu_addresses_df.sample(1).iloc[0]
    pickup_fake = Faker(random.choice(eu_locales))
    pickup_payload = {
        "shipment_id": shipment_response.json()['id'],
        "stop_number": 1,
        "type": "pickup",
        "name": pickup_fake.company(),
        "description": f"Pickup location - {pickup_fake.company_suffix()}",
        "address_line": pickup_address['address_line'],
        "postal_code": pickup_address['postal_code'],
        "city": pickup_address['city'],
        "country_code": pickup_address['country_code'],
        "reference": f"PU-{pickup_fake.bothify('???-###')}",
        "weight": 500.0,
        "weight_unit": "kg",
        "volume": 2.5,
        "volume_unit": "m3",
        "pallet_count": 2,
        "start_datetime": "2024-01-15T08:00:00Z",
        "end_datetime": "2024-01-15T10:00:00Z",
        "comment": f"Loading dock {random.randint(1, 10)}",
        "notes": {
            "contact_person": pickup_fake.name(),
            "phone": pickup_fake.phone_number(),
            "loading_equipment": random.choice(["forklift required", "crane available", "manual loading"])
        }
    }

    # Select random delivery address from real EU addresses (different from pickup)
    delivery_address = eu_addresses_df.sample(1).iloc[0]
    delivery_fake = Faker(random.choice(eu_locales))
    delivery_payload = {
      "shipment_id": shipment_response.json()['id'],
      "stop_number": 2,
      "type": "delivery",
      "name": delivery_fake.company(),
      "description": f"Delivery location - {delivery_fake.company_suffix()}",
      "address_line": delivery_address['address_line'],
      "postal_code": delivery_address['postal_code'],
      "city": delivery_address['city'],
      "country_code": delivery_address['country_code'],
      "reference": f"DL-{delivery_fake.bothify('???-###')}",
      "weight": random.randint(500, 1350),
      "weight_unit": "kg",
      "volume": 2.5,
      "volume_unit": "m3",
      "pallet_count": 2,
      "start_datetime": "2024-01-15T14:00:00Z",
      "end_datetime": "2024-01-15T16:00:00Z",
      "comment": f"Call {random.randint(15, 60)} minutes before arrival",
      "notes": {
        "contact_person": delivery_fake.name(),
        "phone": delivery_fake.phone_number(),
        "unloading_instructions": random.choice(["Use rear entrance", "Loading dock available", "Street level delivery"])
      }
    }

    response = requests.post(url, json=pickup_payload)
    print(f"Pickup stop response for shipment {shipment_response.json()['id']}: {response.status_code} - {response.text}")

    response = requests.post(url, json=delivery_payload)
    print(f"Delivery stop response for shipment {shipment_response.json()['id']}: {response.status_code} - {response.text}")
```
